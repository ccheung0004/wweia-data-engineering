```python
energy = energy_p/energy_total*100
protein = protein_p/protein_total*100
carb = carb_p/carb_total*100
sugar = sugar_p/sugar_total*100
fiber = fiber_p/fiber_total*100
totalFat = totalFat_p/totalFat_total*100
satFat = satFat_p/satFat_total*100
vitA = vitA_p/vitA_total*100
thiamin = thiamin_p/thiamin_total*100
riboflavin = riboflavin_p/riboflavin_total*100
niacin = niacin_p/niacin_total*100
folate = folate_p/folate_total*100
choline = choline_p/choline_total*100
vitC = vitC_p/vitC_total*100
vitE = vitE_p/vitE_total*100
calc = calcium_p/calcium_total*100
mag = mag_p /mag_total*100
iron = iron_p/iron_total*100
pot = pot_p/pot_total*100
sod = sodium_p/sodium_total*100
```

```python
createNutrientData = {
    'Nutrient': ['Energy', 'Protein', 'Carbohydrate', 'Added sugars', 'Dietary fiber', 'Total fat',
                 'Saturated fat', 'Vitamin A', 'Thiamin', 'Riboflavin', 'Niacin', 'Folate (DFE)',
                 'Choline', 'Vitamin C', 'Vitamin E', 'Calcium', 'Magnesium', 'Iron', 'Potassium', 'Sodium'],
    'Percent': [energy, protein, carb, sugar, fiber, totalFat, satFat, vitA, thiamin, riboflavin, niacin,folate, choline, vitC, vitE, calc, mag, iron, pot, sod]
}
```

```python
nutrientDf = pd.DataFrame(createNutrientData)
```

```python
nutrientDf = nutrientDf.sort_values('Percent', ascending = True)
```

```python
ax = nutrientDf.plot.barh(x='Nutrient', y='Percent', legend=False, color='blue', edgecolor='black')

plt.xlabel('Percent')
plt.title('Percentage of daily intakes of energy and selected nutrients contributed by sweet foods among children')
plt.tight_layout()

plt.show()
```
